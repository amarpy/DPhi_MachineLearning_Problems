```python
# create a correlation matrix to see the correlation of variables with the target and among themselves
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
house_data_corr = house_data.corr(method = "pearson")
sns.heatmap(house_data_corr, cmap="magma_r", annot=True)
```

```python
# Add a new column to show age of the house as of 2015 and drop the 'yr_built' variable
# We can remove "yr_renovated" feature as most of its values are 0 as shown in the distplot
# We may also remove 'zipcode' as it is a categorical variable.
# We may also remove'lat', 'long' for this problem, as they are geo coordinates (we can use these if we classify lat and long into zones)
# 'view' may also be removed as they is highly skewed
# 'sqft_living', 'sqft_lot', 'sqft_above', 'grade', 'sqft_lot15', 'sqft_living15' are all strongly correlated with each other. We will keep only 'sqft_living' among these as it is more correlated with the target variable

house_data["age"] = 2015 - house_data["yr_built"]

house_data = house_data.drop(columns = ["yr_built", "yr_renovated", "zipcode", "lat", "long", "view", "sqft_lot", "sqft_above", "grade", "sqft_lot15", "sqft_living15", "waterfront"])
```

```python
X, y = house_data.drop(columns = ["price"]), house_data["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
y_train_pred = lr_model.predict(X_train)

r2 = r2_score(y_train, y_train_pred)
rmse = mean_squared_error(y_train, y_train_pred, squared=False)

print(r2)
print(rmse)
```

```python
y_test_pred = lr_model.predict(X_test)

r2 = r2_score(y_test, y_test_pred)
rmse = mean_squared_error(y_test, y_test_pred, squared=False)

print(r2)
print(rmse)
```
